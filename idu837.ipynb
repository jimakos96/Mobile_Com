```python
import datetime
import pandas as pd
from math import *
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
import itur
import astropy.units as u
```

```python

def global_map(k):
    # -*- coding: utf-8 -*-
    """ This example shows how to compute the rainfall-rate (mm/hr) exceeded
    for 0.01 % of the time of the average year over a large region of the Earth.

    This image is similar to the one plotted in page 5 of Recommendation
    ITU-R P.837-7.
    """


    # Set Recommendation ITU-R P.837 to version 7
    itur.models.itu837.change_version(k)

    # Generate a regular grid of latitude and longitudes with 0.1 degree resolution
    # for the region of interest.
    lat, lon = itur.utils.regular_lat_lon_grid(resolution_lat=0.1,
                                               resolution_lon=0.1)

    # Compute the rainfall rate exceeded for 0.01 % of the time.
    p = 0.01
    R001 = itur.models.itu837.rainfall_rate(lat, lon, p)

    # Display the results in a map
    fig = plt.figure(figsize=(16, 20))
    ax = fig.add_subplot(1, 1, 1)
    m = itur.utils.plot_in_map(
        R001, lat, lon, cmap='jet', vmin=0, vmax=90, ax=ax,
        cbar_text='Rainfall rate exceeded for 0.01% of an average year [mm/hr]')
    return
```

```python
global_map(6)
```

```python
global_map(7)
```

```python
def fun3(lat,lon,k):
    # percentage probability of rain in an average year
    Rp=[]  #rainfall rate exceeded for p% of an average year  (mm/h)
    j=[]
    itur.models.itu837.change_version(k)
    for i in arange(0.01,1,0.001):  #  μεταβάλλω   το χρονικό ποσοστό του μέσου έτους  για να πάρω δείγματα 
        x1=itur.models.itu837.rainfall_rate(lat,lon, i)
        Rp.append(x1)
        j.append(i)
    a=Rp*(u.mm/u.h)
    
    return j,a



```

```python
itur.models.itu837.change_version(7)
x1=itur.models.itu837.rainfall_probability(38,28)
x1 # overall probability of rain throught the year in my region from version x
```

```python
#3h askisi
#a)erwthma
 
#python itur library same results as itur_p837_5.m matlab file 
X1,Y1=fun3(38,28,6)
X2,Y2=fun3(37.98722 ,23.72750,6)
X3,Y3=fun3(39.66667,20.85,6)

plt.figure(figsize=(20,10))
plt.title("Diagram according to ITU837_6 created 2012")
plt.plot(X1,Y1, 'r',label="My Region (Beydağ, Turkey)")  
plt.plot(X2,Y2, 'b',label="Athens")  
plt.plot(X3, Y3, 'g',label="Iwannina")
plt.legend(loc="upper right")
plt.xlabel("exceedance probability p%")
plt.ylabel("Rain fall rate Rp [mm/hr] ")
plt.grid()
plt.show()
```
